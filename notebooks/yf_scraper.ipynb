```python
from yahooquery import Ticker
import pandas as pd
import time
```

```python
tickers = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA', 'META', 'NFLX', 'NVDA', 'AMD', 'BA', 
           'V', 'SPY', 'SPCE', 'FB', 'TWTR', 'BABA', 'MSTR', 'DIS', 'PYPL', 'SHOP', 'COIN', 
           'SQ', 'INTC', 'CSCO', 'IBM', 'GE', 'WMT', 'T']
data_list = []
```

```python
stock = Ticker(tickers)  # Fetch all at once
stats = stock.key_stats
```

```python
if not isinstance(stats, dict):
    print("Error: Unexpected format in stats:", stats)
    exit()
```

```python
data_list = []
for ticker in tickers:
    stat = stats.get(ticker, {})  # Get stats for the ticker or an empty dict
    
    if isinstance(stat, str):  # If the API returns an error message
        print(f"Error fetching data for {ticker}: {stat}")
        continue
    
    data = {
        "Ticker": ticker,
        "Price/Book": stat.get("priceToBook", "N/A"),
        "EV/EBITDA": stat.get("enterpriseToEbitda", "N/A"),
    }
    data_list.append(data)
```

```python
df = pd.DataFrame(data_list)
```
